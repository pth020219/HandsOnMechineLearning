```python
# 최종 파라미터. C = 0.000192, gamma = 0.2847, degree=2, kernel="Poly", coef0 = 10

ovr_svc = Pipeline([
    ("scaler", StandardScaler()),
    ("ovr_svc", OneVsRestClassifier(SVC(C=0.000192, gamma=0.2847, degree=2, kernel="poly", coef0=10), n_jobs=-1))
])
```

```python
# 모델 훈련

ovr_svc.fit(X_train, y_train)
```

```python
from sklearn.metrics import f1_score

final_prediction = ovr_svc.predict(X_test)

f1 = f1_score(y_test, final_prediction, average="micro")
print(f1)
```
