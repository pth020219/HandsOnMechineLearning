```python
# 선형적으로 분리되는 데이터 셋에 LinearSVC를 훈련시켜 보세요. 그런 다음 같은 데이터셋에 SVC와 SGDClassifier를 적용해보세요.
# 거의 비슷한 모델이 만들어 지는지 확인해 보세요.
```

```python
import numpy as np
from sklearn import datasets
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC, SVC
from sklearn.linear_model import SGDClassifier
from sklearn.utils import shuffle
import matplotlib.pyplot as plt
```

```python
#선형 데이터셋 준비

iris = datasets.load_iris()
X = iris["data"][:, (2, 3)]
y = (iris["target"] == 2).astype(np.float64)

X, y = shuffle(X, y, random_state=42)
```

```python
# 3 가지 모델을 위한 파이프 라인 구성

linsvc = Pipeline([
    ("scaler", StandardScaler()),
    ("linsvc", LinearSVC(C=1, loss="hinge", random_state=42))
])

svc = Pipeline([
    ("scaler", StandardScaler()),
    ("svc", SVC(C=1, kernel="linear", random_state=42))
])

sgdclf = Pipeline([
    ("scaler", StandardScaler()),
    ("sgd_clf", SGDClassifier(alpha=0.0067, loss="hinge", eta0=0.0001, random_state=42))
])

```

```python
linsvc.fit(X, y)

svc.fit(X, y)

sgdclf.fit(X, y)
```

```python
print("the parameters of LinearSVC : ")
print(linsvc.named_steps["linsvc"].coef_)
print(linsvc.named_steps["linsvc"].intercept_)

print("the parameters of SVC : ")
print(svc.named_steps["svc"].coef_)
print(svc.named_steps["svc"].intercept_)

print("the parameters of SGDClassifier : ")
print(sgdclf.named_steps["sgd_clf"].coef_)
print(sgdclf.named_steps["sgd_clf"].intercept_)
```

```python
# 각 결정 경계의 기울기와 편향을 계산합니다
w1 = -linsvc.named_steps["linsvc"].coef_[0, 0]/linsvc.named_steps["linsvc"].coef_[0, 1]
b1 = -linsvc.named_steps["linsvc"].intercept_[0]/linsvc.named_steps["linsvc"].coef_[0, 1]
w2 = -svc.named_steps["svc"].coef_[0, 0]/svc.named_steps["svc"].coef_[0, 1]
b2 = -svc.named_steps["svc"].intercept_[0]/svc.named_steps["svc"].coef_[0, 1]
w3 = -sgdclf.named_steps["sgd_clf"].coef_[0, 0]/sgdclf.named_steps["sgd_clf"].coef_[0, 1]
b3 = -sgdclf.named_steps["sgd_clf"].intercept_[0]/sgdclf.named_steps["sgd_clf"].coef_[0, 1]

# 결정 경계를 원본 스케일로 변환합니다
line1 = linsvc.named_steps["scaler"].inverse_transform([[-10, -10 * w1 + b1], [10, 10 * w1 + b1]])
line2 = svc.named_steps["scaler"].inverse_transform([[-10, -10 * w2 + b2], [10, 10 * w2 + b2]])
line3 = sgdclf.named_steps["scaler"].inverse_transform([[-10, -10 * w3 + b3], [10, 10 * w3 + b3]])

# 세 개의 결정 경계를 모두 그립니다
plt.figure(figsize=(11, 4))
plt.plot(line1[:, 0], line1[:, 1], "k:", label="LinearSVC")
plt.plot(line2[:, 0], line2[:, 1], "b--", linewidth=2, label="SVC")
plt.plot(line3[:, 0], line3[:, 1], "r-", label="SGDClassifier")
plt.plot(X[:, 0][y==1], X[:, 1][y==1], "bs") # label="Iris-Virginica"
plt.plot(X[:, 0][y==0], X[:, 1][y==0], "yo") # label="non Iris-Virginica"
plt.xlabel("Petal length", fontsize=14)
plt.ylabel("Petal width", fontsize=14)
plt.legend(loc="upper center", fontsize=14)
plt.axis([0, 5.5, 0, 5])

plt.show()
```
